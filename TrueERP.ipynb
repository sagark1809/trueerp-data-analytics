```python
#ERP Data Analytics on TrueERP Purchase Bills Data
This Notebook performs data cleaning, data filtering, and data analysis on purchase bill data exported from the TrueERP system using Python Pandas
```

```python
import pandas as pd
```

```python
#Data Loading: The Dataset ised in this analysis is a CSV export of purchase bills from the TrueERP system.

df = pd.read_csv("AllPurchaseBills.csv")
```

```python
df.head()
```

```python
#Data Understanding: This step helps to understand the structure, data types, and overall quality of the ERP dataset.
df.info()
df.describe()
```

```python
#Data Cleaning: Data Cleaning is performed to handle misiing values, correct data types, and standardize column names to make the dataset analysis-ready.

#check missing values
df.isna().sum()
```

```python
# Invoice Number is a mandotory field in ERP System.

#handle missing invoice numbers
df["Invoice No"] = df["Invoice No"].fillna("UNKNOWN")
```

```python
#Invoice date is converted into datetime format to enable time-based analysis.

#Convert Invoice Date to Date Format
df["Invoice Date"] = pd.to_datetime(df["Invoice Date"], dayfirst=True)
```

```python
#Column names are standardized for easier referencing and consistency.

#Standaedize Column Names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("()","",regex=False)
)
df.columns
```

```python
#Data Filtering is applied to extract meaningful subsets of data based on business conditions.

#data filtering


#identify high balue purchase bills
high_value_bills = df[df["total_amount_"] > 50000]
high_value_bills[["supplier_name", "total_amount_"]]
```

```python
#supplier specific filtering

df[df["supplier_name"] == "MAHARASHTRA HARDWARE"]
```

```python
#Data Analysis: This Section performs analytical operations to derive insights from the cleaned and filtered ERP data

#Total Purchase, Paid, and Oustanding Amount
total_purchase = df["total_amount_"].sum()
total_paid = df["paid_amount_"].sum()
total_balance = df["balance_amount_"].sum()

total_purchase, total_paid, total_balance
```

```python
#Monthly Purchase Trend Analysis
monthly_purchase = (
    df.groupby(df["invoice_date"].dt.to_period("M"))["total_amount_"]
    .sum()
    .reset_index()
)

monthly_purchase
```

```python
#Top 10 Suppliers by Purchase Value
top_suppliers = (
    df.groupby("supplier_name")["total_amount_"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_suppliers
```

```python
#Payment Effeciency Analysis
df["payment_ratio"] = df["paid_amount_"] / df["total_amount_"]
df[["supplier_name", "payment_ratio"]].head()
```

```python
#TDS Analysis
total_tds = df["tds"].sum()
total_tds
```

```python
#Export Cleaned Dataset: The Clean Dataset is reported for reporting and dashboard creation.

df.to_csv("cleaned_purchase_bills.csv", index=False)
```
